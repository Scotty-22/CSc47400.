```python
import numpy as np
import matplotlib.pyplot as plt

x = data['FL']
y1 = data['RW']
y2 = data['CW']

plt.figure(figsize=(9,3))

plt.subplot(1,2,1)
plt.scatter(x, y1, color='blue')
plt.xlabel('FL')
plt.ylabel('RW', color='blue')
plt.twinx()
plt.scatter(x, y2, color='red', marker='x')
plt.ylabel('CW', color='red')
plt.title('A) FL-RW & FL-CW')

groups = np.array([f"{sp.decode() if isinstance(sp, bytes) else sp}{sx.decode() if isinstance(sx, bytes) else sx}"
                   for sp, sx in zip(data['sp'], data['sex'])])

colors = {'BM':'blue','BF':'cyan','OM':'red','OF':'orange'}

plt.subplot(1,2,2)
for g in colors:
    mask = (groups == g)
    plt.scatter(data['FL'][mask], data['RW'][mask], color=colors[g], label=g)

plt.xlabel('FL')
plt.ylabel('RW')
plt.title('B) Groups')
plt.legend(fontsize=8)

plt.tight_layout()
plt.savefig('Figure_MultiPanel_Simple.svg', bbox_inches='tight')
plt.show()



```
